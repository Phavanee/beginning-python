```python
import pandas as pd
import sklearn as sk
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import numpy as np
```

### Series

```python
x = pd.Series([11,3,2,4,23])
x
```

```python
x = pd.Series([11,3,2,4,23],index=['r1','w','h','e',100])           # like hash keys???
x
```

```python
print(x['r1'], x.r1, x[100], x.iloc[4])                      #iloc is integer location
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}            # wow this is just like hash keys ;; curly brackets
mydata = pd.Series(calories)
mydata
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
mydata = pd.Series(data)
mydata
```

### DataFrame
