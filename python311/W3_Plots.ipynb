### subplot

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.subplot(2,2,1)
plt.pie(x, labels=mylabels, radius=0.8)
plt.subplot(2,2,2)
plt.bar(x=[1,2,3,4], height=x, width=0.8)
plt.subplot(2,2,3)
plt.plot(x)
plt.subplot(2,2,4)
plt.scatter(np.random.rand(10),np.random.rand(10))
plt.savefig('test1.png')
```

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.pie(x, labels=mylabels, radius=0.8)
plt.subplot(2,2,2)
plt.bar(x=[1,2,3,4], height=x, width=0.8)
plt.subplot(2,2,3)
plt.plot(x)
plt.subplot(2,2,4)
plt.scatter(np.random.rand(10),np.random.rand(10))
plt.savefig('test2.png')
```

```python
x = 4 + np.random.normal(0, 2, 100)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(1, 255, len(x))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
ax1.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
ax2.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
fig.savefig('test3.png')
```

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
ax2.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)
```

```python
# plot
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig = plt.figure()

x = [0, 1, 2]

data1 = [1, 2, 3]
data2 = [1, 4, 9]

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x, data1)
ax2.plot(x, data2)

plt.show()
```
